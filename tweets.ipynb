## Import Library

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import csv
```

## Get Tweets with snscrape

```python
query = "ibu kota negara (ikn) lang:id until:2022-02-28 since:2021-10-01" #keyword and filters
tweets = [] #declaration
limit = 5000 #nums of tweet

#shout out to ai_spectrumüôèüèª
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.content])
        
df_tweets = pd.DataFrame(tweets, columns=['date', 'tweet'])
```

```python
df_tweets
```

```python
df_tweets.drop_duplicates(subset='tweet', inplace=True)
```

## Export to .csv

```python
df_tweets['date'] = pd.to_datetime(df_tweets["date"].dt.strftime('%Y-%m-%d'))
df_tweets
```
